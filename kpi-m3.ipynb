```python
import requests
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import pylab
pylab.rcParams['figure.figsize']=16,12
pylab.rcParams['figure.dpi'] = 300
```

```python
payload = '{"query": [{"code":"Tid", "selection":{"filter":"all","values":["*"]}}],"response": {"format": "json"}}'
url = 'http://api.scb.se/OV0104/v1/doris/sv/ssd/PR/PR0101/PR0101A/KPIFFMP'
url2 = 'http://api.scb.se/OV0104/v1/doris/sv/ssd/START/PR/PR0101/PR0101A/KPISkuggM'
r1 = requests.post(url1,data=payload)
r2 = requests.post(url2,data=payload)
```

```python
r1.encoding='utf-8-sig'
dict = r1.json()['data']
inflation = pd.DataFrame.from_dict(dict)
inflation['values']=inflation['values'].apply(lambda value: float(value[0]))
inflation['date']=inflation['key'].apply(lambda key: dt.datetime.strptime(str(key[0]),"%YM%m"))
inflation = inflation.set_index('date')
inflation = inflation.drop('key',1)
inflation.columns=['inflation']
```

```python
r2.encoding='utf-8-sig'
dict = r2.json()['data']
kpi = pd.DataFrame.from_dict(dict)
kpi['values']=kpi['values'].apply(lambda value: float(value[0]))
kpi['date']=kpi['key'].apply(lambda key: dt.datetime.strptime(str(key[0]),"%YM%m"))
kpi = kpi.set_index('date')
kpi = kpi.drop('key',1)
kpi.columns = ['kpi']
```

```python
moneySupply = pd.read_excel('http://www.scb.se/Statistik/FM/FM5001/2014M12/FM5001_2014M12_DI_01_SV_penningmangd.xls',
                          sheetname=2,skiprows=7,parse_cols='A,C,E')
```

```python
moneySupply['date'] = moneySupply.apply(lambda row: dt.date(int(row[0]),int(row[1]),1),axis=1)
moneySupply=moneySupply.set_index('date')
moneySupply=moneySupply.drop(moneySupply.columns[[0,1]],axis=1)
```

```python
frame = (moneySupply.join(inflation)).join(kpi)
```
